# MẢNG HAI CHIỀU
- Trong Python để sử dụng mảng 2 hay nhiều chiều ta sử dụng nested list. Ở phần mảng 2 chiều này các bạn sẽ học về các bài toán cơ bản trên mảng 2 chiều như bài toán về hàng, cột, các phép toán trên ma trận, ma trận xoáy ốc..
- Để giải quyết các bài toán liên quan tới ma trận các bạn cũng sử dụng nested list trong Python

## 1. Tạo mảng 2 chiều và nhập xuất:
- Để khai báo mảng 2 chiều, ta cần chỉ rõ số lượng hàng, cột của ma trận.

```python
n, m = 2, 3
a = [[0 for _ in range(m)] for _ in range(n)]
for i in range(n):
    a[i] = list(map(int, input().split()))
for i in range(n):
    for j in range(m):
        print(a[i][j], end = " ")
    print()
```

## 2. Truy cập vào các phần tử trong mảng 2 chiều:
- Để truy cập vào các phần tử trong mảng, các bạn dùng chỉ số hàng, và chỉ số cột. Chỉ số và Cột của mảng 2 chiều được đánh số từ 0 tương tự như mảng 1 chiều.


## 3. Các bài toán trên mảng 2 chiều:
### a. Tìm phần tử lớn nhất, nhỏ nhất trong mảng:


```python
a = [[1,2,5], [3,1,0], [4,1,5]]
min_val , max_val =  10**18 , -10**18
for row in a:
    min_val = min(min_val, min(row))
    max_val = max(max_val, max(row))
print(min_val, max_val)
```

### b) Flatten mảng 2 chiều thành mảng 1 chiều:


```python
a = [[1,2,5], [3,1,0], [4,1,5]]
b = [x for small_list in a for x in small_list]
print(min(b))
print(max(b))
print(b)
```

### c) Tính tổng từng hàng của ma trận (mảng 2 chiều):


```python
a = [[1,2,5], [3,1,0], [4,1,5]]
sum_row = [sum(row) for row in a]
print(sum_row)
```

## 4. Các phép toán trên ma trận:
### a) Tạo ma trận chuyển vị: 
- Ma trận chuyển vị của ma trận a cỡ nxm là ma trận b có cỡ mxn, trong đó các hàng của ma trận a trở thành các cột của ma trận b.

```python
a = [[1,2,5],
[3,1,0],
[4,1,5]]
a_t = [[row[i] for row in a] for i in range(len(a[0]))]
print(a_t)
```

### b) Cộng trừ 2 ma trận:
- Trong đại số tuyến tính, ma trận tương tự như một mảng 2 chiều gồm n hàng và m cột. Để 2 ma trận có thể cộng hoặc trừ cho nhau thì chúng phải có cùng số hàng và số cột.


### c) Nhân hai ma trận: 
- Giả sử có 2 ma trận a cỡ nxm, ma trận b cỡ pxq, để ma trận a có thể nhân với ma trận b thì số cột của ma trận a, tức là m phải bằng số hàng của ma trận b, tức là p.

- Khi đó m = p thì ma trận tích của a và b sẽ là ma trận c có cỡ nxq. Phần tử ở chỉ số (i,j) của ma trận tích c được tính bằng cách nhân từng cặp phần tử ở hàng i của ma trận a với các phần tử ở cột j của ma trận b.

```python
n,m,p = map(int,input().split())
a, b = [0] * n, [0] * m
for i in range(n):
    a[i] = list(map(int,input().split()))
for i in range(m):
    b[i] = list(map(int,input().split()))
c = [[0 for _ in range(n)] for _ in range(p)]
for i in range(n):
    for j in range(p):
        for k in range(m):
            c[i][j] += a[i][k] * b[k][j]
print(c)
```

## 5. Kỹ thuật duyệt các ô liền kề:


```python
a = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
path = [[-1, 0], [0,-1],[0,1],[1,0]]
i, j = 1, 1 # 5
for x in path:
    i1, j1 = i + x[0], j + x[1]
    print(a[i1][j1], end = " ")
```
