## 2. Else:
- Else được sử dụng trong trường hợp condition bên trong if là sai

```python
if condition:
    #code if condition is True
else:
    #code if condition is False
```

```python
n = 100
if n%2 == 0:
    print("Even")
else:
    print("Odd")
```

## 3.Elif
- Từ khóa (else if) trong python được sử dụng bên dưới if để kiểm tra thêm điều kiên bổ sung nếu điều kiện bên trên sai. Các điều kiện ở bên trong if và elif nếu đúng thì khối code tương ứng sẽ được thực thi, nếu không có điều kiện nào đúng thì khối lệnh bên trong else được thực thi.

```python
if condition1:
    #code1
elif condition2:
    #code2
....
elif condition3:
    #codeN
else:
    #codeElse
```

```python
a, b = 100, 200
if a < b:
    print(a, 'less than', b)
elif a == b:
    print(a,'equal to', b)
else:
    print(a,'greater than', b)
```

## 4. Shorthand if và toán tử ba ngôi
### shorthand if:
- shorthand if: Bạn có thể sử dụng câu lệnh if trên 1 dòng

```python
a, b = 100, 200
if a < b : print(a, 'less than', b)
```

- Nếu trong if có nhiều câu lệnh, bạn có thể đặt dấu chấm phẩy giữa các câu lệnh

```python
a, b = 100, 200
if a < b : print(a, 'less than', b ); print('Luccute'); print('abcd')
```

### Toán tử ba ngôi:


```python
a, b = 100, 200
res = 'Luccute' if a < b else 'python'
print(res)
```

## 5.If lồng nhau
- Khi điều kiện trong if quá phức tạp, bạn có thể sử dụng if lồng nhau (nested if) để kiểm tra từng điều kiện một

```python
# Kiểm tra N xem có phải là số nằm trong đoạn [20,30] và chia hết cho ít nhất một trong ba số 2, 3, 5. Nếu đúng in ra YES, Ngược lại in ra NO
# Cách 1
n = 27
if n >= 20 and n <= 30 and (n%2 == 0 or n %3 == 0 or n % 5 == 0):
    print("YES")
else:
    print("NO") 
```

```python
# Cách 2
n = 27
if n >= 20 and n <= 30:
    if n % 2 == 0 or n % 3 == 0 or n % 5 == 0:
        print("YES")
    else:
        print("NO")
else:
    print("NO")
```
