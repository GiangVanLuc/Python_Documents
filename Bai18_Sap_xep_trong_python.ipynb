## 1.Hm sort() ca list:
- Khi hm sort thiu cc tham s ny n hot ng bnh thng.

```python
a = [5, 1, 3 ,2 , 4]
a.sort()
print(a)
b = ["python", "Luc", "c++", "java"]
b.sort()
print(b)
```

- Sort list theo th t gim dn.


```python
a = [5, 1, 3, 2, 4]
a.sort(reverse = True)
print(a)
b = ["python", "luccute","cute","java"]
b.sort(reverse = True)
print(b)
```

- Custom sort vi tham s key

```python
# V d: Sp xp theo tng ch s tng dn.
def sumDigit(n):
    res = 0
    while n != 0:
        res += n % 10
        n //= 10
    return res

if __name__ == "__main__":
    a = [3,24, 111, 30, 12, 21, 99, 32]
    a.sort(key = sumDigit)
    print(a)
```

## 2. Hm sort() vi lambda:
- tham s key ca hm sort c th s dng biu thc lambda  thay th.


```python
# Sort theo tr tuyt i
a = [-10, 230, 5, -4, -3, 10]
a.sort(key = lambda x: abs(x))
print(a)
a.sort(key = lambda x: abs(x), reverse = True)
print(a)
```

```python
# sort nested list:
a = [[1,2], [3,1],[3,4],[0,5],[1,1]]
a.sort(key = lambda x : x[0])
print(a)
```

- Tham s key ca hm sort c th s dng biu thc lambda  thay th.

```python
# Sort nested list theo th t tng dn ca thnh phn th nht v gim dn theo thnh phn th 2:
a = [[1,2],[3,1],[3,4],[0,5],[1,3]]
a.sort(key = lambda x : (x[0], -x[1]))
print(a)
```

```python
# sort mt list cc dict: 
a = [
    {'name' : 'Giang Van Luc', 'job' : 'Dev', 'salary' : 5000},
    {'name' : 'Nguyen Van sec', 'job' : 'Dev', 'salary' : 1500},
    {'name' : 'Le Van Cuong', 'job' : 'tester', 'salary' : 500},
    {'name' : 'NGuyen Van Khieu', 'job' : 'BA', 'salary' : 2000}
]
a.sort(key = lambda x : x.get('salary'))
for x in a:
    print(x)
```

## 3. Kt hp hm sort vi itemgetter() v attrgetter():
- Bn c th s dng itemgetter() v attrgetter() t module operator.


```python
from operator import itemgetter
friends = [("Huynh Manh Tuong", 30),
           ("Tran Nhat Phi",28),
           ("Tuan Binh",21),
           ("Thieu Tuan",35)]
friends.sort(key = itemgetter(1))
for x in friends:
    print(x)
```

```python
# sort theo nhiu tiu ch
from operator import itemgetter
friends = [("Huynh Manh Tuong", 30,'B'),
           ("Tran Nhat Phi",28,'A'),
           ("Tuan Binh",28,'C'),
           ("Thieu Tuan",30,'Z')]
friends.sort(key = itemgetter(1,2))
for x in friends:
    print(x)
```

## 4. Xy dng hm so snh cho hm sort bng cmp_to_key trong functools


```python
# comparison function: comparator
from math import *
from functools import cmp_to_key
# -1  1  0
# nu a l thng ng trc , b l thng ng sau sau khi ma bn sp xp 
# nu a v b  ng th t m bn mun th tr v  -1, ngc li cc bn tr v 1
def cmp(a,b):
    if a < b :
        return - 1
    elif a > b:
        return 1
    else:
        return 0

a = [10, 2, 3, 1, 4, 5, 0]
a.sort(key = cmp_to_key(cmp))
print(a)
```

## 5. Thut ton tm kim nh phn - Binary search


```python
def Binary_search(a,l,r,x):
    while(l <= r):
        m = (l+r)//2
        if(a[m] == x):
            return True
        elif (a[m] < x):
            l = m + 1
        else:
            r = m - 1
    return False

if __name__ == "__main__":
    a = [1,1,2,2,3,4,5,6]
    if Binary_search(a,0,len(a)-1 , 6):
        print("OK")
    else:
        print("NOT")
```
