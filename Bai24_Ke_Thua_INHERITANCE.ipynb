# 4. Subclassing:
- Thng thng lp con s c theem nhng thuc tnh mi so vi lp cha, tuy nhin  trnh d tha code, bn c th gi hm khi to ca lp cha trong hm khi to ca lp con  khi to cc thuc tnh m lp con k tha t lp cha.


```python
class Person:
    def __init__(self, name, birth):
        self.name = name
        self.birth = birth
class Student(Person):
    def __init__(self, name, birth, gpa):
        Person.__init__(self, name, birth)
        self.gpa = gpa
    def display(self):
        print(f'{self.name} {self.birth} {self.gpa:.2f}')

if __name__ == "__main__":
    s = Student('Luccute', '09/04/2005',2.5)
    s.display()
```

# 5. Ghi  hm:
- Trong trng hp lp con v lp cha c phng thc cng tn, khi  bn ang ghi  phng thc ny
- Thng thng phng thc ny  lp con s c ci t mt cch chi tit hn v c th gi phng thc tng ng  lp cha  trnh d tha code.

```python
# Lp cha:
class Person:
    def __init__(self, name, birth):
        self.name = name
        self.birth = birth
    def show(self):
        return f'{self.name} {self.birth}'
```

```python
# Lp con
class Student(Person):
    def __init__(self, name, birth, gpa):
        super().__init__(name,birth)
        self.gpa = gpa
    def show(self):
        return Person.show(self) + " " + f'{self.gpa:.2f}'
```

```python
if __name__ == '__main__':
    t = Person('Nam', '22/12/2005')
    print(t.show())
    s = Student('Luccute', '09/04/2005', 2.5)
    print(s.show())
```

# 6. Cc kiu k tha:
a k tha - Multiple Inheritance:
- Mt lp c th cng k tha nhiu lp khc nhau c gi l a k tha.


```python
# V d:
class A:
    def __init__(self, a):
        self.a = a
class B:
    def __init__(self, b):
        self.b = b
class C(A, B):
    def __init__(self, a, b):
        A.__init__(self,a)
        B.__init__(self, b)
    def show(self):
        print(self.a, self.b)
c = C(100, 200)
c.show()
```

- Khi lp con li c mt lp con khc k tha n ta c k tha nhiu mc.

```python
# V d:
class A:
    def __init__(self, a):
        self.a = a
class B(A):
    def __init__(self, b):
        self.b = b
class C(B):
    def __init__(self, a, b):
        A.__init__(self,a)
        B.__init__(self, b)
    def show(self):
        print(self.a, self.b)
c = C(100, 200)
c.show()
```
