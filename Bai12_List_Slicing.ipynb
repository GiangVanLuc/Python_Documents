# LIST SLICING
- Pythong list slicing l mt k thut gip cc bn c th truy cp vo 1 khong cc phn t trong list thng qua ton t


## 1. Slicing vi ch s khng m


```python
a = [2, 3, 1, 5, 7, 4, 3]
b = a[0 : 4 : 1]
print(b)
c = a[0 : 4]
print(c)
```

- Nu khng c tham s stop th mc nh gi tr ca stop s l ch s cui cng trong mng, nu khng c tham s start th gi tr mc nh ca start s l 0

```python
a = [2, 3, 1, 5, 7, 4, 3]
b = a[3: ]
print(b)
c = a[ : 4]
print(c)
```

```python
# Bc nhy l 2:
a = [2, 3, 1, 5, 7, 4, 3]
b = a[0: 5 : 2]
print(b)
```

## 2. Slicing vi ch s m


```python
# V d
a = [2, 3, 1, 5, 7, 4, 3]
b = a[-4 : - 1]
print(b)
```

```python
# Kt hp ch s m v ch s dng:
a = [2, 3 ,1, 5, 7, 4, 3]
b = a[1 : -2]
print(b)
```

## 3. Lt ngc list:
-  lt ngc list cc bn ch cn b trng phn start v stop, phn step c gi tr l - 1.


```python
a = [2, 3, 1, "Pythong", "Luccute"]
b = a[::-1]
print(b)
```

## 4. Thay i gi tr ca nhiu phn t trong mt on
- Bn c th thay i hoc xa 1 lot cc phn t trong 1 on xc nh bng list slicing. Ch  cc bn ch c th gn 1 iterable cho slice m bn ct ra.


## 5. Chn v xa vi list slicing:


```python
a = ['x', 'y','z']
a[:0] = [10, 20, 30]
print(a)
```

```python
a = ['x', 'y','z']
a[len(a):] = [10, 20, 30]
print(a)
```

```python
a = ['x', 'y','z']
a[1:1] = [100, 200, 300]
print(a)
```

```python
a = ['x', 'y','z','t']
a[1:2] = [100, 200, 300]
print(a)
```

```python
a = ['x','y','z','t','u','v']
a[1:4] = []
print(a)
```

## 6. Copy mt list (shallow copy):


```python
a = ['x','y','z','t','u','v']
b = a[:]
print(b)
print(b == a)
print(b is a)
```
