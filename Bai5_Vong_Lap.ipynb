## 1 Vòng lặp for và hàm range()
### a) Vòng lặp for
- thay vì duyệt qua các giá trị số thì Python sẽ lặp trên các iterable (list, tuple, string, set,..). Thứ tự duyệt sẽ theo thứ tự xuất hiện trong iterable. Để thực hiện vòng lặp for ta sử dụng built-in function là range()



### b) Hàm range()
- Hàm range() sẽ sinh ra một dãy số và bạn sẽ sử dụng vòng for để duyệt qua từng số trong dãy đã sinh ra.


```python
# this is a example

#start = 0, stop = 5, step = 1

for i in range(5):
    print('Luccute', end = ' ')
```

```python
# start = 0, stop = 5, step = 1
for i in range(0,5,1):
    print("Luccute", end = ' ')
```

```python
# start = 2, end = 5, step = 1
for i in range(2,5):
    print("LUCCUTE", end = ' ')
```

```python
# start = 1, stop = 11, step = 2

for i in range(1,11,2):
    print(i, end = " ")
```

```python
# in ra các ước của N

n = 100
for i in range(1,100,1):
    if n % i == 0:
        print(i, end = ' ')
```

### c) vòng for lồng nhau
- Vòng for lồng nhau (Nested loop) xuất hiện khi một câu lệnh bên trong vòng for này lại là một vòng for khác. Để hiểu được cách hoạt động của vòng for các bạn chỉ cần nhớ rằng mỗi vòng lặp của vòng for bên ngoài thì toàn bộ vòng for con bên trong sẽ được thực hiện

```python
for i in range(3):
    print("Vong for ngoai khi i =", i)
    for j in range(2):
        print(i,j)
```

## 2. Câu lệnh break và continue
### a) câu lệnh break:
- Câu lệnh break được sử dụng để kết thúc vòng lặp ngay lập tức, vòng lặp for sẽ kết thúc ngay tại thời điểm gặp câu lệnh break và tiếp tục các câu lệnh bên dưới vòng for. Thông thường thì câu lệnh break sẽ đi kèm theo một điều kiện kích hoạt

```python
for i in range(5):
    print(i, end =' ')
    if i == 3: break
    print('28teck', end = ' ')
print('Ket thuc')
```

### b) Câu lệnh continue:
- Câu lệnh continue được dùng để bỏ qua lần lặp hiện tại và quay trở lại luôn vòng lặp tiếp theo. Các câu lệnh bên dưới continue ở trong vòng lặp sẽ được bỏ qua.

```python
for i in range(5):
    print('Luccute')
    continue
    print("Python")
```

## 3. Vòng lặp while
### a) Vòng lặp while, break, continue:
- While được sử dụng khi bạn muốn thực hiện một tác vụ vô thời hạn, cho đến khi một điều kiện cụ thể được đáp ứng. Đây là một vòng lặp được kiểm soát theo điều kiện và thường sử dụng dụng khi chưa biết trước số lượng vòng lặp

```python
# cú pháp
while condition:
    # code when condition is True
else:
    # code While condition is False
```

```python
# this is a example
n = 1
while n < 5:
    print(n)
    n += 1
print('Following statement')
```

```python
n = 1
while n < 5: 
    print(n)
    n += 1
else:
    print("End")
print("FOLLOWING statement")
```

```python
# Tính tổng chữ số của số nguyên

n = 12345
sum_digit = 0
while n != 0:
    sum_digit += n % 10
    n //= 10
print("SUM: ", sum_digit)
```

```python
# Lật ngược một số nguyên

n = 12345
rev = 0
while n != 0:
    rev = rev * 10 + n % 10
    n //= 10
print("REV: ", rev)
```

```python
# Câu lệnh break sẽ làm else trong while không được thực hiện

i = 1
while i <= 4:
    print(i)
    if i == 3: break
    i += 1
else:
    print("END")
print("FOLLOWING statement")
```

```python
# Câu lệnh continue

i = 1
while i <= 4:
    print("LUCCUTE")
    i += 1
    continue
    print("PYTHON")
else:
    print("END")
print("FOLLOWING STATEMENT")
```

### b) Vòng lặp vĩnh viễn
- Để tạo 1 vòng lặp vĩnh viễn thì bạn chỉ cần để điều kiện là True

```python
# Vòng lặp sẽ lặp vĩnh viễn cho tới khi bạn nhập xâu là 28tech

while True:
    str = input("ENTER YOUR PASSWORD: ")
    if str == 'LUCCUTE':
        print("OK")
        break
    else:
        print("NOT OK")
```

```python
# YÊU CẦU nhập lại khi người dùng nhập 1 số nguyên âm hoặc số 0

n = -1
while True:
    n = int(input("Enter a positive number: "))
    if n > 0:
        break
print(n)
```
