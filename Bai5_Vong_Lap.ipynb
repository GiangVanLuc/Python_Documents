## 1 Vng lp for v hm range()
### a) Vng lp for
- thay v duyt qua cc gi tr s th Python s lp trn cc iterable (list, tuple, string, set,..). Th t duyt s theo th t xut hin trong iterable.  thc hin vng lp for ta s dng built-in function l range()



### b) Hm range()
- Hm range() s sinh ra mt dy s v bn s s dng vng for  duyt qua tng s trong dy  sinh ra.


```python
# this is a example

#start = 0, stop = 5, step = 1

for i in range(5):
    print('Luccute', end = ' ')
```

```python
# start = 0, stop = 5, step = 1
for i in range(0,5,1):
    print("Luccute", end = ' ')
```

```python
# start = 2, end = 5, step = 1
for i in range(2,5):
    print("LUCCUTE", end = ' ')
```

```python
# start = 1, stop = 11, step = 2

for i in range(1,11,2):
    print(i, end = " ")
```

### c) vng for lng nhau
- Vng for lng nhau (Nested loop) xut hin khi mt cu lnh bn trong vng for ny li l mt vng for khc.  hiu c cch hot ng ca vng for cc bn ch cn nh rng mi vng lp ca vng for bn ngoi th ton b vng for con bn trong s c thc hin

```python
for i in range(3):
    print("Vong for ngoai khi i =", i)
    for j in range(2):
        print(i,j)
```

## 2. Cu lnh break v continue
### a) cu lnh break:
- Cu lnh break c s dng  kt thc vng lp ngay lp tc, vng lp for s kt thc ngay ti thi im gp cu lnh break v tip tc cc cu lnh bn di vng for. Thng thng th cu lnh break s i km theo mt iu kin kch hot

```python
for i in range(5):
    print(i, end =' ')
    if i == 3: break
    print('28teck', end = ' ')
print('Ket thuc')
```

### b) Cu lnh continue:
- Cu lnh continue c dng  b qua ln lp hin ti v quay tr li lun vng lp tip theo. Cc cu lnh bn di continue  trong vng lp s c b qua.

```python
for i in range(5):
    print('Luccute')
    continue
    print("Python")
```

## 3. Vng lp while
### a) Vng lp while, break, continue:
- While c s dng khi bn mun thc hin mt tc v v thi hn, cho n khi mt iu kin c th c p ng. y l mt vng lp c kim sot theo iu kin v thng s dng dng khi cha bit trc s lng vng lp

```python
# c php
while condition:
    # code when condition is True
else:
    # code While condition is False
```

```python
# this is a example
n = 1
while n < 5:
    print(n)
    n += 1
print('Following statement')
```

```python
n = 1
while n < 5: 
    print(n)
    n += 1
else:
    print("End")
print("FOLLOWING statement")
```

```python
# Tnh tng ch s ca s nguyn

n = 12345
sum_digit = 0
while n != 0:
    sum_digit += n % 10
    n //= 10
print("SUM: ", sum_digit)
```

```python
# Lt ngc mt s nguyn

n = 12345
rev = 0
while n != 0:
    rev = rev * 10 + n % 10
    n //= 10
print("REV: ", rev)
```

```python
# Cu lnh break s lm else trong while khng c thc hin

i = 1
while i <= 4:
    print(i)
    if i == 3: break
    i += 1
else:
    print("END")
print("FOLLOWING statement")
```

```python
# Cu lnh continue

i = 1
while i <= 4:
    print("LUCCUTE")
    i += 1
    continue
    print("PYTHON")
else:
    print("END")
print("FOLLOWING STATEMENT")
```
