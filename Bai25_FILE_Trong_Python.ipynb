# FILE TRONG PYTHON

- Python hỗ trợ các thao tác phổ biến với file, bao gồm tạo xóa, đọc ghi file.

## 1. Hàm open():
- Để có thể thao tác với file, trước hết các bạn cần mở file này lên, khi mở file cần tên file và mode mở.


- Ngoài ra bạn có thể chỉ rõ đang làm việc với file text hay file nhị phân bằng các cách thêm "t" (file text) hoặc "b" (file nhị phân) vào cuối mode mở file.

## 2. Đọc file:
- Sau khi mở file xong bạn có thể sử dụng hàm read() để đọc toàn bộ thông tin trong file text.

## 3. Tạo và ghi file:
- Khi mở file với mode "w" và "a" sẽ tạo file mới nếu file chưa tồn tại và làm việc với file cũ nếu nó tồn tại.


- Để xóa file ta sử dụng hàm remove() của module os.
