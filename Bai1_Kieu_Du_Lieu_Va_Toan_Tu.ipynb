## 1. Câu lệnh print

- Giúp in ra nội dung lên màn hình
- Trước khi được in ra màn hình, các đối tượng(object) này được chuyển thành xâu kí tự

print(object, sep = seperator, end = end)
-----------------------------------------
Các tham số của hàm print :
- object: các đối tượng trong python
- sep: Phân cách giữa các đối tượng khi in, nếu khong có tham số này thì mặc định sẽ là dấu cách
- end: Chỉ ra kí tự được in ở cuối của dòng, nếu không có tham số này thì mặc định sẽ là dấu xuống dòng

```python
print("Hello world","Luccute", sep = "-", end = "///" )
```

## 3. Biến và kiểu dữ liệu

### 3.1 Biến:
- Biến là vùng chứa để lưu trữ dữ liệu phục vụ bài toán
- Trong Python các bạn khong cần phải khai báo biến cũng như chỉ rõ kiểu dữ liệu của biến đó, biến sẽ được tạo và xác định kiểu tự động (dynamic typing) khi bạn gán giá trị cho nó
- Để biết kiểu dữ liệu của biến các bạn có thể sử dụng hàm type()

```python
# Example
a = 100
print(type(a))
s = 'Luccute'
print(type(s))
```

### Chú ý khi sử dụng biến
- Không được đặt tên biến có chưa dấu cách, kí tự đặc biệt, bắt đầu bằng chữ số
- Tên biến trong Python phân biệt hoa thường (case sensitive)

### 3.2 Kiểu dữ liệu:
- Kiểu dữ liệu: Trong Python có 3 kiểu dữ liệu số: 
 1. Số nguyên (integer)
 2. Số thực dấu phẩy động (Floating- point numbers) 
 3. Số Phức (Complex Numbers)

```python
# Example
a = 100
b = 5.2
c = 5 + 3j
print(type(a),type(b),type(c),sep = '\n')
```

#### a) Số nguyên
- Đối với số nguyên thì trong Python không có giới hạn về giá trị mà số nguyên có thể lưu, bạn có thể xử lý số nguyên lớn với Python

```python
a = 32532532532532532532532532532532553253252
print(type(a))
print(a)
```

- Trong Python thì các số nguyên thường được in ra dưới dạng cơ số 10 nhưng bạn cũng có thể in ra các số ở hệ 2,8,16.
- Cơ số 2 (Binary) ->
 Tiền Tố: '0b' hoặc '0B'
- Cơ số 8 (Octal) ->
 Tiền Tố: '0o' hoặc '0O' 
- Cơ số 16 (Hexadecimal) ->
 Tiền Tố: '0x' hoặc '0X' 

```python
# this is a example
a = 0b1101
print(a)
b = 0o123
print(b)
c = 0x22A
print(c)
```

### b) Số thực
- số thực là số âm, dương kèm theo phần thập phân
- Trong Python giá trị số thực có thể lưu lớn nhất xấp xỉ 1.8*10^308, các giá trị lớn hơn giá trị này được mô tả bởi chuỗi inf (Infinity). Giá trị số thực nhỏ nhất có thể lưu là 5.0*10^-324, các giá trị nhỏ hơn số này được coi là 0.

```python
a = 3.5
print(a)
b = 3e5
print(b)
d = 1.9e308
print(d)
e = 5.4e-325
print(e)
```

### c) Số Phức
- số phức gồm phần thực (real part) và phần ảo (imaginary part) đi kèm j
- bạn có thể in ra phần thực phần ảo của số phức x bằng x.real và x.imag

```python
a = 3 + 5j
print(a.real)
print(a.imag)
```

### f) Ép kiểu:
- Khi bạn muốn chỉ định kiểu cho một biến nào đó bạn có thể ép kiểu cho nó. Quá trình casting trong Python được hoàn thành bằng cách sử dụng construcor

- str(): Ép kiểu sang string


```python
a = str(230)
b = str(3234.44)
c = str(True)
print(a, b ,c)
```

## 4. Toán tử:
### Các loại Toán tử trong Python

- Toán tử Gán : Assignment operator
- Toán tử toán học: Arithmetic operator
- Toán tử so sánh: Comparision operator
- Toán tử logic: Logical operator
- Toán tử nhận dạng: Identity operator
- Toán tử thành viên: Membership operator
- Toán tử bit: Bitwise operator

### a) Toán tử gán:
- CÚ PHÁP : a = b
-> Ý nghĩa: Gán giá trị của b cho a

```python
a = 100
b = a
print(b)
```

```python
# Gán giá trị cho nhiều biến trong cùng 1 câu lệnh
a, b , c = 100, 200, 300
print(a, b , c)
```

```python
# Hoán vị giá trị của 2 biến trong python
a = 'Luccute'
b = 'Python'
a, b = b, a
print(a, b)
```

### b) Toán tử toán học


```python
# this is a example
a = 100 + 200
b = 100 - 200
c = 100 * 200
d = 100 / 200
e = 300 // 200
f = 15 % 2
g = 2**10
print(a,b,c,d,e,f,g,sep = '\n')
```

### c) Toán tử so sánh
- Khi bạn sử dụng toán tử so sánh thì kết quả của phép so sánh sẽ trả về True hoặc False


### d) Toán Tử logic:
- Bạn có thể sử dụng các toán tử and, or, not để kết hợp nhiều biểu thức so sánh


### e) Toán tử nhận dạng:
- Toán tử nhận dạng được sử dụng để so sánh 2 đối tượng chứ không phải so sánh 2 giá trị


```python
a = [1, 2, 3]
b = [1, 2, 3]

print(a == b)
print(a is b)
print(id(a))
print(id(b))
```

### f) Toán tử thành viên:
- Toán tử thành viên được dùng để kiểm tra sự tồn tại của một đối tượng trong list, xâu, tuple.
