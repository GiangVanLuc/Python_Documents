# NHẬP INPUT

## Nhập dữ liệu với hàm input:
- Cú pháp: input(prompt)
- Giá trị trả về: input() trả về xâu kí tự ở kiểu str, các bạn cần chú ý ép kiểu sang dữ liệu tương ứng của biến trong đề bài.

```python
# this is a example

n = input('Nhap so :')
print(type(n))
print(n)
```

```python
# ép sang kiểu là số nguyên hoặc thực bằng hàm int(), float()

n = int(input('Nhap so: '))
print(type(n))
print(n)
```

```python
# this is a example
# Nhập 2 số trên một dòng

a, b = map(int, input().split())
print(a, b)
```

```python
# Nhập 4 số trên một dòng

a, b, c, d = map(int, input().split())
print(a, b , c, d)
```
