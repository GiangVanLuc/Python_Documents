# MAP
## 1. Map():
- Map() là một hàm trong python có chức năng apply một hàm có sẵn cho mọi phân tử trong 1 iterable (list, tuple, str,...)


- Apply một hàm tự định nghĩa

```python
def add(n):
    return n + 100

if __name__ == "__main__":
    a = range(5)
    b = list(map(add, a))
    print(b)
```

- Apply một hàm built - in

```python
if __name__ == "__main__":
    s = "LUC94"
    b = list(map(ord, s))
    print(b)
```

- Đọc 1 mảng số nguyên từ input và chuyển vào list

```python
if __name__ == "__main__":
    a = list(map(int, input().split()))
    print(a)
```


- Trong trường hợp hàm apply trong map ngắn gọn bạn có thể thay thế bằng lambda

```python
def square(n):
    return n ** 2

if __name__ == "__main__":
    a = [1, 2, 3, 5]
    b = list(map(square, a))
    print(b) 
    c = list(map(lambda x : x ** 2 , a))
    print(c)
```

- Áp dụng map với nhiều iterable:


```python
def add(a, b):
    return a + b

if __name__ == "__main__":
    a = [1, 2, 3, 4]
    b = [2, 3, 5 , 4]
    c = list(map(add,a , b))
    print(c)
```

- Áp dụng map với nhiều iterable sử dụng trong lambda:

```python
if __name__ == "__main__":
    a = [1, 2, 3, 4]
    b = [2, 3, 5, 4]
    c = list(map(lambda x, y : x + y, a, b))
    print(c)
```

## 2. Filter()
- Filter() được sử dụng để trích xuất các phần tử trong một iterable khi apply một hàm nào đó với phần tử đó mà hàm trả về giá trị là True.


```python
# lọc ra các số chẵn trong list
def even(n):
    return n % 2 == 0

if __name__ == "__main__":
    a = [1, 2, 3, 4, 5]
    b = list(filter(even,a))
    print(b)
```

```python
# Lọc ra các số nguyên tố trong list
import math
def prime(n):
    for i in range(2, math.isqrt(n) + 1):
        if n % i == 0:
            return False
    return n > 1

if __name__ == "__main__":
    a = [20, 10, 17, 7, 29, 19]
    b = list(filter(prime,a))
    print(b)
```

```python
# Áp dụng filter với lambda
if __name__ == "__main__":
    a = [20, 10, 17, 7, 29, 19]
    b = list(filter(lambda x : x >= 15, a))
    print(b)
    
```
