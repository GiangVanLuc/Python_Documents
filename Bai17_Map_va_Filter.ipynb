# MAP
## 1. Map():
- Map() l mt hm trong python c chc nng apply mt hm c sn cho mi phn t trong 1 iterable (list, tuple, str,...)


- Apply mt hm t nh ngha

```python
def add(n):
    return n + 100

if __name__ == "__main__":
    a = range(5)
    b = list(map(add, a))
    print(b)
```

- Apply mt hm built - in

```python
if __name__ == "__main__":
    s = "LUC94"
    b = list(map(ord, s))
    print(b)
```

- c 1 mng s nguyn t input v chuyn vo list

```python
if __name__ == "__main__":
    a = list(map(int, input().split()))
    print(a)
```


- Trong trng hp hm apply trong map ngn gn bn c th thay th bng lambda

```python
def square(n):
    return n ** 2

if __name__ == "__main__":
    a = [1, 2, 3, 5]
    b = list(map(square, a))
    print(b) 
    c = list(map(lambda x : x ** 2 , a))
    print(c)
```

- p dng map vi nhiu iterable:


```python
def add(a, b):
    return a + b

if __name__ == "__main__":
    a = [1, 2, 3, 4]
    b = [2, 3, 5 , 4]
    c = list(map(add,a , b))
    print(c)
```

- p dng map vi nhiu iterable s dng trong lambda:

```python
if __name__ == "__main__":
    a = [1, 2, 3, 4]
    b = [2, 3, 5, 4]
    c = list(map(lambda x, y : x + y, a, b))
    print(c)
```

## 2. Filter()
- Filter() c s dng  trch xut cc phn t trong mt iterable khi apply mt hm no  vi phn t  m hm tr v gi tr l True.


```python
# lc ra cc s chn trong list
def even(n):
    return n % 2 == 0

if __name__ == "__main__":
    a = [1, 2, 3, 4, 5]
    b = list(filter(even,a))
    print(b)
```

```python
# Lc ra cc s nguyn t trong list
import math
def prime(n):
    for i in range(2, math.isqrt(n) + 1):
        if n % i == 0:
            return False
    return n > 1

if __name__ == "__main__":
    a = [20, 10, 17, 7, 29, 19]
    b = list(filter(prime,a))
    print(b)
```

```python
# p dng filter vi lambda
if __name__ == "__main__":
    a = [20, 10, 17, 7, 29, 19]
    b = list(filter(lambda x : x >= 15, a))
    print(b)
    
```
